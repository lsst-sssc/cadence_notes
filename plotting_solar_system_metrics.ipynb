```python
cols = metrics['sso']
cols_lc=metrics['Lightcurve Inversion']
number_of_sims_per_plot=9

#plot the discovery metrics 
for r in version:
    print(r)
    S =summary[r][cols] / summary['1.7'][cols].loc[baseline] 
    S.columns = short_names['sso']
    T =summary[r][cols_lc] / summary['1.7'][cols_lc].loc[baseline] 
    T.columns = short_names['Lightcurve Inversion']
    startindex=0
    endindex=number_of_sims_per_plot

    for i in np.arange(int(np.round(len(S)/np.float(number_of_sims_per_plot)+0.5)), dtype=int):
        # plot detection fractions
        D=S[startindex:endindex]
        startindex=endindex
        if (endindex+number_of_sims_per_plot>len(S)):
            endindex=len(S)
        else:
            endindex=endindex+number_of_sims_per_plot

        xi = np.arange(len(D))
    
        xnames = D.index.values
        D.plot(figsize=(20, 9), style=styles['sso'])
        plt.legend(loc=(1, 0.05))
        plt.xticks(xi, xnames, rotation=90, fontsize='large')
        plt.title('Discovery Variation (%)')
        plt.grid(True, alpha=0.3, color='k', linestyle=':')
        plt.tight_layout()
        plt.savefig('discovery_completeness_all_v'+str(r)+'_'+str(np.int(i))+'.pdf')
        plt.close('all')
        
     # plot lc metics 
    
    startindex=0
    endindex=number_of_sims_per_plot
    
    for i in np.arange(int(np.round(len(T)/np.float(number_of_sims_per_plot)+0.5)), dtype=int):
        # plot detection fractions
        D=T[startindex:endindex]
        startindex=endindex
        if (endindex+number_of_sims_per_plot>len(T)):
            endindex=len(T)
        else:
            endindex=endindex+number_of_sims_per_plot

        xi = np.arange(len(D))
    
        xnames = D.index.values
        D.plot(figsize=(20, 9), style=styles['sso'])
        plt.legend(loc=(1, 0.05))
        plt.xticks(xi, xnames, rotation=90, fontsize='large')
        plt.title('Light Curve Inversion')
        plt.grid(True, alpha=0.3, color='k', linestyle=':')
        plt.tight_layout()
        plt.savefig('light_curve_inversion_all_v'+str(r)+'_'+str(np.int(i))+'.pdf')
        plt.close('all')

        
        
```
